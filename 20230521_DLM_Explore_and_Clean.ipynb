#Dates

It is known that some test orders are put in many days, weeks or months before the specimen is collected. I'll limit this data set to only those specimens collected in the specified time period.

First I need to check if the data is in the correct format and convert it if necessary.

```python
dlmdata5.dtypes
```

The Date Time columns are not in the correct format. It takes a long time for the computer to change each column's format, so it's best to choose those that really need to be changed.

Collection Date Time may be best as this is when labels are printed for samples, which have presumably just been drawn.

Receiving Date Time should be soon after that and may be a better option in case tubes are sometimes pre-labelled.

```python
dlmdata5[['Test Name', 'Result Name','Result Text',
       'Result Cancelled', 'Ultimately Cancelled', 'Ordered Date Time',
       'Collection Date Time', 'Receiving Date Time', 'Resulted Date Time',
       'Verify Date Time']].head(10)
```

```python
dlmdata6 = dlmdata5.copy()
dlmdata6['Collection Date Time'] = pd.to_datetime(dlmdata6['Collection Date Time'])
dlmdata6['Receiving Date Time'] = pd.to_datetime(dlmdata6['Receiving Date Time'])
```

It might make sense to exclude the remaining tests that were never received, but first need to find out why they have a -nul- result instead of a (.) result.

```python
excldata1 = dlmdata6.copy()
excldata1 = excldata1[(excldata1['Collection Date Time'] >='2022-06-01 00:00:00') | (excldata1['Collection Date Time'] <'2022-03-01 00:00:00')]
excldata2 = excldata1[['Test Name', 'Result Name', 'Order Number','Modified Result','Result Text','Result Cancelled', 'Collection Date Time', 'Ordered Date Time', 'Receiving Date Time',
                      'Resulted Date Time', 'Verify Date Time']]
excldata2.head(15)
```

Some tests are definitely given labels long before the sample is collected. Up to a year!!

```python
excldata3 = dlmdata6.copy()
excldata3 = excldata3[(excldata3['Receiving Date Time'] >='2022-06-01 00:00:00') | (excldata3['Receiving Date Time'] <'2022-03-01 00:00:00')]
excldata4 = excldata3[['Test Name', 'Result Name', 'Order Number','Modified Result','Result Text','Result Cancelled', 'Collection Date Time', 'Ordered Date Time', 'Receiving Date Time',
                      'Resulted Date Time', 'Verify Date Time']]
excldata2.head(15)
```

Receiving Date seems a much better option. I will allow a receiving date from Feb 23 to accommodate batch testing.

Clearly the extract was limited to all verified by end of May as nothing came up after that date.
